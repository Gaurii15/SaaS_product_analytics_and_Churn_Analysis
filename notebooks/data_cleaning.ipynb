# Build the Master Analytics Dataset

Merge Accounts + Subscriptions

```python
master_df = accounts.merge(
    subscriptions,
    on="account_id",
    how="left"
)
```

```python
master_df.shape
master_df.head()
```

```python

master_df.shape
```

```python
master_df = master_df.merge(
    feature_usage,
    on="subscription_id",
    how="left"
)
```

master_df.shape
master_df.head()

Merge Support Tickets

```python
master_df = master_df.merge(
    support,
    on="account_id",
    how="left"
)
```

Merge Churn Events

```python
master_df = master_df.merge(
    churn,
    on="account_id",
    how="left"
)
```

```python
master_df.shape
master_df.columns
```

```python
master_df.shape
```

### Feature Engineering

Aggregate the data by account
- Create Feature Usage Metrics

```python
feature_metrics = master_df.groupby("account_id").agg(
    total_feature_usage=("usage_count", "sum"),
    avg_usage_duration=("usage_duration_secs", "mean"),
    total_errors=("error_count", "sum"),
    beta_feature_usage=("is_beta_feature", "sum")
).reset_index()
```

Create Support Metrics

```python
support_metrics = master_df.groupby("account_id").agg(
    total_tickets=("ticket_id", "count"),
    avg_resolution_time=("resolution_time_hours", "mean"),
    avg_first_response=("first_response_time_minutes", "mean"),
    avg_satisfaction=("satisfaction_score", "mean"),
    escalation_count=("escalation_flag", "sum")
).reset_index()
```

Create Revenue Metrics

```python
revenue_metrics = master_df.groupby("account_id").agg(
    total_mrr=("mrr_amount", "sum"),
    total_arr=("arr_amount", "sum")
).reset_index()
```

Create Subscription Metrics

```python
subscription_metrics = master_df.groupby("account_id").agg(
    plan_type=("plan_tier_y", "first"),
    billing_frequency=("billing_frequency", "first"),
    auto_renew=("auto_renew_flag", "first")
).reset_index()
```

Customer Information
From accounts table

```python
customer_info = accounts[[
    "account_id",
    "industry",
    "country",
    "seats",
    "is_trial",
    "churn_flag"
]]
```

#### Final Analytics Dataset

```python
analytics_df = customer_info.merge(feature_metrics, on="account_id", how="left")
analytics_df = analytics_df.merge(support_metrics, on="account_id", how="left")
analytics_df = analytics_df.merge(revenue_metrics, on="account_id", how="left")
analytics_df = analytics_df.merge(subscription_metrics, on="account_id", how="left")
```
